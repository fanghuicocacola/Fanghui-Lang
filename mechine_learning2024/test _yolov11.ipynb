```python
from ultralytics import YOLO

if __name__ == '__main__':
    # 加载模型
    model = YOLO(model=r'yolo11n.pt')
    # 进行推理
    model.predict(source=r'trains\wuyanzu.jpg',     # source是要推理的图片路径这里使用yolo自带的图片
                    save=True,    # 是否在推理结束后保存结果
                    show=False,    # 是否在推理结束后显示结果
                    project='runs',  # 结果的保存路径
                    )
    
```
