```python
import numpy as np
import os
import re
```

```python
data = {
    'docs':[],
    'labels':[]
}
```

```python
os.listdir('email')
```

```python
os.listdir('email/spam')
```

```python
for label in os.listdir('email'):
    for filename in os.listdir('email/'+label):
        doc = open('email/'+label+'/'+filename,errors= 'ignore').read()
        # 数据切分
        doc = [tok.lower() for tok in re.split(r'\W', doc) if len(tok) > 1]
        data['docs'].append(doc)
        data['labels'].append(label)
```

```python
def getDocList(docs):
    # 初始化空集合
    docSet = set([])
    for doc in  docs:
        # 取交集
        docSet = set(doc) | docSet
    docList = list(docSet)
    docList.sort()
    return docList
```

```python
docList = getDocList(data['docs'])
```

```python
def doc2V(doc,docList):
    # 单词集合 的 维度
    dims = len(docList)
    # 初始化 坐标点
    doc_v = np.zeros(dims)

    for word in doc:
        if word in docList:
            doc_v[docList.index(word)] += 1
    # 文档 ——> 向量
    return doc_v
```

```python
docs_v  = list(map(lambda doc:doc2V(doc,docList),data['docs']))
```

```python
def train(docs_v,labels,docList):
    # Abusive
    n_doc = len(labels)
    # 向量基 永远是 docList
    docLen = len(docList)

    # 防止零概率出现
    p1num = np.ones(docLen)
    p0num = np.ones(docLen)
    p1Denum,p0Denum = 2, 2
    # 遍历 所有的 doc 向量
    for i in range(n_doc):
        if labels[i] == 'spam':
            p1num += docs_v[i]
            p1Denum += np.sum(docs_v[i])
        elif labels[i] == 'ham':
            p0num += docs_v[i]
            p0Denum += np.sum(docs_v[i])
    # p1doc, p0doc, pA
    return np.log(p1num/p1Denum), np.log(p0num/p0Denum),labels.count('spam')/n_doc
    
p1doc,p0doc,pA = train(docs_v,data['labels'],docList)
p1doc
```

```python
def classify(doc,p1doc,p0doc,pA):
    doc_v  = doc2V(doc,docList)
    # P(doc|C1) = np.sum(np.log(doc_v * p1doc)) * P(C1)
    p1 = np.sum(doc_v * p1doc) + np.log(pA)
    p0 = np.sum(doc_v * p0doc) + np.log(1-pA)

    if p1 > p0:
        return 'spam'
    else:
        return 'ham'
```

```python
classify(data['docs'][11],p1doc,p0doc,pA)
```
