```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
test = pd.read_table('horseColicTest.txt',header = None)
test.head()
```

```python
train = pd.read_table('horseColicTraining.txt',header = None)
train.head()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def BGD_LR(df,alpha=0.01,maxCycles=500):
    xMat = np.mat(df.iloc[:,:-1])
    yMat = np.mat(df.iloc[:,-1])

    m,_ = xMat.shape
    xMat = np.column_stack((np.ones(m), xMat))

    # 样本的行和列
    m,n = xMat.shape
    weights = np.zeros((n,1))
    # 最优化算法： 迭代更新

    for i in range(maxCycles):
        h = sigmoid(xMat * weights)
        grad = xMat.T * (h - yMat.T)/m
        weights -= alpha*grad
    return weights
```

```python
def classifyVector(inX):
    if inX > 0.5:
        return 1.0
    else:
        return 0.0
```

```python
def colicTest():
    # 训练集 =》 weights
    ws = BGD_LR(train,alpha=0.001,maxCycles=5000)
    # xMat默认为1
    xMat = np.array(test.iloc[:,:-1])
    xMat = np.column_stack((np.ones(xMat.shape[0]),xMat))
    # 预测标签
    predict = sigmoid(np.array(np.mat(xMat) * np.mat(ws)).flatten())
    predict = [ classifyVector(i) for i in predict]
    Acc = np.mean(test.iloc[:,-1] == predict)
    return Acc
```

```python
colicTest()
    
```
