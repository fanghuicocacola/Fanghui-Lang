```python
df_1 = pd.read_table('lenses.txt',header=None)
df_1.columns = ['age','prescript','astigmatic','tearRate','label']
```

```python
df_1
```

```python
def dfMap(df):
    for col in df_1.columns[:-1]:
        col_values = list(set(df[col]))
        d = dict(zip(col_values, list(range(len(col_values)))))
        df[col] = df[col].map(d)
    return df
```

```python
dfMap(df_1)
```

```python
def create_plot(df):
    clf = DecisionTreeClassifier(criterion="entropy",max_leaf_nodes=30,random_state=0)
    x_train = np.array(df.iloc[:, :-1])
    y_train = np.array(df.iloc[:, -1])
    clf = clf.fit(x_train,y_train)
    return clf
```

```python
clf = create_plot(df_1)
```

```python
col_name = list(df_1.columns[:-1])
labels = list(set(df_1.label))
```

```python
plt.figure(figsize=(200,90))
tree.plot_tree(clf,filled=True,feature_names=col_name,class_names=labels)
```
