```python
import numpy as np
import pandas as pd
import os
```

```python
def getFile(dir):
    files = os.listdir(dir)
    imgs = []
    labels = []
    for file_name in files:
        # 数字文件 -》 一行字符串
        imgs.append(''.join(np.loadtxt(f'{dir}/{file_name}',dtype=str)))
        # 提取数字标签
        labels.append(file_name.split('_')[0])
    return pd.DataFrame({
        'imgs':imgs,
        'labels':labels
    })
```

```python
def getFiles(dirs):
    return list(map(getFile,dirs))
```

```python
[test, train] = getFiles(["../testDigits","../trainingDigits"])
```

```python
def hanming(A,B):
    return sum([a != b for (a,b) in zip(A,B)])
```

```python
def knn(inX, df, k):
    dist = df.iloc[:,0].apply(lambda x: hanming(inX,x))

    dist_1 = pd.DataFrame({
        'dist': dist,
        'label': df.iloc[:, -1]
    })

    dist_k = (dist_1.sort_values(by='dist')).iloc[:k]
    return dist_k.value_counts('label').index[0]
```

```python
def digitsTest(test,train,k):
    predict = []
    for __, row in test.iterrows():
        predict.append(knn(row[0],train,k))
    return np.mean(predict == test.iloc[:,-1])    
```
